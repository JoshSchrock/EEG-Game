Set up DMD - reference https://www.sciencedirect.com/science/article/pii/S0165027015003829

```python

matricies = []
h = 200
length = Xs.shape[2]-h
for X in Xs:
    Xaug = X[:, 0:length]
    print(Xaug.shape)
    for k in range(1, h):
        Xaug = np.concatenate((Xaug, X[:, k:length+k]), axis=0)
        print(Xaug.shape)
    matricies.append(Xaug)
    print(Xaug.shape)


# print(X[0])
# dmd = DMD()
# dmd.fit(Xaug)
```
